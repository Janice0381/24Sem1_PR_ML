## Final Selected Model: LightGBM

```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold, GridSearchCV

kf = KFold(n_splits=5, shuffle=True, random_state=random_state)
scoring = "neg_mean_squared_error"

models = {}

model = LGBMRegressor(learning_rate=0.1,
                      data_sample_strategy="goss",
                      top_rate=0.2,
                      other_rate=0.1,
                      force_col_wise=True,
                      verbosity=0,
                      n_jobs=-1,
                      random_state=random_state)

# Define the hyperparameters and their possible values
param_grid = {
    "n_estimators": [25, 50],
    "reg_alpha": [0, 0.1],   #regularization term
    "reg_lambda": [0, 0.1],   #regularization term
    "enable_bundle": [True, False]   #bundling
}

grid_search = GridSearchCV(model, param_grid, cv=kf, scoring=scoring, refit=True)
grid_search.fit(X_train, y_train)

print("Best parameters: ", grid_search.best_params_)
print("Best CV score: {:.6f}".format(grid_search.best_score_))

models["LightGBM"] = grid_search.best_estimator_
```

## **Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

for _name, _model in models.items():
    y_pred = _model.predict(X_test)
    print("{:>17}: MSE={:.4f} | MAE={:.4f} | R^2={:.4f}".format(
        _name,
        mean_squared_error(y_test, y_pred),
        mean_absolute_error(y_test, y_pred),
        r2_score(y_test, y_pred)
        )
    )
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

model = DecisionTreeClassifier(criterion="entropy",
                               max_depth=5,
                               min_samples_split=2,
                               min_impurity_decrease=0.0)

model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)
print("Estimated probs:", y_prob[:10])

y_cls = model.predict(X_test)
print("Estimated classes:", y_cls[:10])
print()
```

```python
# Accuracy, F1, and ROC AUC scores
print("Accuracy:", accuracy_score(y_test, y_cls))
print("F1:", f1_score(y_test, y_cls))
print("ROC AUC:", roc_auc_score(y_test, y_prob[:, 1]))
```

### The following is an interpretation of the result of predicting the testset.

```python
'''
From the 6 ensemble methods, LightGBM method showed the best performance with the lowest MSE and the highest R^2 value.


If the string variable has two values, it will be converted to 0 or 1 and if it has more than two, it will be preprocessed and analyzed using dummy values.
When 'yes' is set to 1 and 'no' is set to 0, the f1 score calculated for four different cases of 'poutcome' and 'month' resulted in 0.3 approximately for all cases.
This suggests there might be an issue in the data preprocessing stage.
According to general conventions, 'yes' usually means "positive" or "present," while 'no' means "negative" or "absent."
Thus, it is generally a good practice to convert 'yes' to 1 and 'no' to 0.
However, depending on the characteristics of the data and the purpose of the analysis, this can be reversed.
If 'no' is more significant, 'no' can be set to 1 and 'yes' to 0.
A scenario where 'yes' can be set to 0 and 'no' to 1 is when there is a significant data imbalance.
When there is a large disparity in the ratio of 'yes' to 'no' in the dataset, it can affect the model's performance.
Hence, considering the class ratio while deciding the conversion method can be crucial.


For example, it might be considered to set the more important class to 1 based on the class ratio.
In the 'train.csv' dataset, out of 32950 'y' values, 29238 are 'no' and 3712 are 'yes',
which indicates a significant data imbalance with a ratio of approximately 1:7.8766 for 'yes' to 'no.'

Therefore, we  set 'yes' to 0 and 'no' to 1.


However, in the prediction.csv file, 'yes' should be 1 and 'no' should be 0, so the value was inverted during the process of converting to an integer at the end.
That is, if you check the csv file, you can see that 'yes' is set correctly as 1 and 'no' is set to 0.
'''
```
